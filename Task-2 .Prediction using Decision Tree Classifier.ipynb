```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label=LabelEncoder()
```

```python
iris["Species"]=label.fit_transform(iris["Species"])
```

```python
iris.head()
```

```python
iris["Species"].value_counts()
```

```python
x=iris.drop(["Id","Species"],axis=1)
y=iris["Species"]             
```

```python
x.head()
```

```python
y.head()
```

```python
# Heatmap
plt.figure(figsize=(10,10))
sns.heatmap(iris.corr(),annot=True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print(accuracy_score(y_pred,y_test))
```

# Plotting the tree graph and visualizing graphically

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(25,25))
_=plot_tree(model,filled=True)
```

# Making Prediction

```python
# for making new prediction you need to enter your new data in input section and then run the code for see prediction

input=(5.0,3.6,1.4,0.2)

# changing the input data to a numpy array
input_to_numpy_array=np.asarray(input)


# reshape the data as we are predicting the label for only one instance
reshaped_array=input_to_numpy_array.reshape(1,-1)


prediction=model.predict(reshaped_array)
print(prediction)

if (prediction[0]==0):
    print("Iris-setosa")
elif(prediction[0]==1):
    print("Iris-versicolor")
else:
    print("Iris-virginica")
```
